```python
#first we will check whether the words in excel file are in english word or not 
#then we will proceed to the second part 
import requests
import bs4
import xlrd
from openpyxl import *
```

```python
#this part to clean the excel sheet
import enchant 
d = enchant.Dict("en_US")

path = "Desktop/Book1.xlsx"
wb = load_workbook(path)
ws = wb["Sheet1"]
inputWorkbook = xlrd.open_workbook(path)
inputWorksheet = inputWorkbook.sheet_by_index(0) 
number_of_rows = inputWorksheet.nrows
words_tm= []

for y in range(1,inputWorksheet.nrows):
    if d.check(inputWorksheet.cell_value(y,1)) == True:
        words_tm.append(inputWorksheet.cell_value(y,1))
    #words.append(inputWorksheet.cell_value(y,1))
    elif d.check(inputWorksheet.cell_value(y,1)) == False:
        ws.delete_rows(y+1)

wb.save(path)
#print(words)
```

```python
#this part will find the synonyms from the Thesaurus.com and save it in the excel file
base_url = 'https://www.thesaurus.com/browse/{} '
path = "Desktop/Book1.xlsx"
inputWorkbook = xlrd.open_workbook(path)
inputWorksheet = inputWorkbook.sheet_by_index(0) 
number_of_rows = inputWorksheet.nrows
words= [] #this save the words from excel in the sheet in a form of list
for y in range(1,inputWorksheet.nrows):
    words.append(inputWorksheet.cell_value(y,1))


list_of_list = [] #this will store the synonyms for each word in the form of strings with "," seperator 
for item in words:
    res = requests.get(base_url.format(item))
    soup = bs4.BeautifulSoup(res.text,'lxml')
    t = soup.select(".css-1s3v085.eh475bn1") 
    list_of_synonyms = []
    for item in range(0,len(t)):
        temp = t[item].getText()
        list_of_synonyms.append(temp)
    s = ","
    s = s.join(list_of_synonyms)    
        
    
    list_of_list.append(s)  

list_of_A_list = []
for item in words:
    res = requests.get(base_url.format(item))
    soup = bs4.BeautifulSoup(res.text,'lxml')
    T = soup.select(".css-itvgb.eh475bn1") 
    list_of_antonyms = []
    for item in range(0,len(T)):
        temp = T[item].getText()
        list_of_antonyms.append(temp)
    r = ","
    r = r.join(list_of_antonyms)    
        
    
    list_of_A_list.append(r)
    
wb = load_workbook(path)
ws = wb["Sheet1"]
j = 0
x = 1
for i in range(2,number_of_rows+1):#this will put the words back in excel with respect to the word
    wcell = ws.cell(i,3)
    wcell.value = list_of_list[j]
    j = j+1
    W_cell = ws.cell(i,1)#this will recreate the sr.number list
    W_cell.value = x
    x = x+1

e = 0
for i in range(2,number_of_rows+1):
    wcell = ws.cell(i,12)
    wcell.value = list_of_A_list[e]
    e = e+1


wb.save(path)
```
